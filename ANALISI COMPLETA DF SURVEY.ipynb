```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv = pd.read_csv('Desktop/survey.csv')
csv
```

```python
csv.head(5)
```

```python
csv.shape
```

```python
csv['Age'].min(), csv['Age'].max()  # indicami il numero minimo e max di Age
```

```python
csv.groupby('Age')
```

```python
age_mean = csv.groupby('Age').min() # mi raggruppa il df per age partendo in ordine crescente per age
age_mean.head(10)
```

```python
age_20 = age_mean.loc[20:30, :] # mi raggruppa il df per age ccon condizione 10 - 20
age_20
```

```python
group_age = csv.groupby('Age').mean() # mi raggruppa per anno indicandomi la media degli altri valori
group_age.head(7)
```

```python
group_age2 = csv.groupby('Age').std() # mi raggruppa per anno indicandomi la deviazione standard degli altri valori
group_age2.head()
```

```python
group_age.index # trova gli indici delle medie
```

```python
plt.scatter(x=group_age.index, y=group_age['Weekly Earnings'])
plt.title("Grafico a dispersione della media delle entrate settimanali in funzione dell'età")
plt.show()
```

```python
group_age2.index # trova gli indici della deviazione standard
```

```python
plt.scatter(x=group_age2.index, y=group_age2['Weekly Earnings'])
plt.title("Grafico a dispersione della deviazione standard delle entrate settimanali in funzione dell'età")
plt.show()
```

```python
plt.scatter(x=csv.Age, y=csv['Weekly Earnings'], c=csv['Age'], alpha=0.04)
plt.title("Grafico a dispersione delle entrate settimanali in funzione dell'età")
plt.show()
```

```python
plt.scatter(x=csv.Age, y=csv['Food & Drink Prep'], c=csv['Age'], alpha=0.07)
plt.title("Grafico a dispersione del tempo della preparazione del cibo in funzione dell'età")
plt.show()
```

```python
csv['Actual_age'] = csv['Year'].apply(lambda year: 2018 - year)
csv
```

```python
csv.sort_values('Actual_age')
```

```python
plt.scatter(x=csv.Actual_age, y=csv['Food & Drink Prep'], c=csv['Age'], alpha=0.07)
plt.title("Grafico a dispersione del tempo della preparazione del cibo in funzione dell'età")
plt.show()
```
